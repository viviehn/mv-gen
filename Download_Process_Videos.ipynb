```python
data_csv = "video_data.csv"
vid_dir = "split_vids_dir"
start_at = 0  # edit this value
global_counter = start_at
split_length = 10  # seconds



output_dir = os.getcwd() + '/' + vid_dir
input_dir = os.getcwd() + '/' + download_dir + '/'

# make vid_dir if not already exists
!mkdir {output_dir}

downloaded_vids = [f for f in os.listdir(input_dir) if os.path.isfile(input_dir + f)]

with open(data_csv, mode="w") as file:
    data_writer = csv.writer(file)
    for vid in downloaded_vids:
        print(vid)
        vid_abs_path = input_dir + vid
        num_vids = split_video(vid_abs_path, output_dir, global_counter, split_length)
        
        split_vid_title = vid.split('-')
        title = split_vid_title[1].split('.')[0]
        artist = split_vid_title[0]
        
        data_writer.writerow([title, artist, global_counter, global_counter + num_vids - 1])
        
        global_counter += num_vids
        
# now process_dir should contain all the split videos
# create csv for youtube-8m
```
